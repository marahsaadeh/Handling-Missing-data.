```python
#find the percentage NAN in each column
def percnt_missing(df):
    percnt_missing_data =(df.isna().sum()*100/row_number)/100
    percentage_NAN_in_each_column=percnt_missing_data.to_dict()
    return percentage_NAN_in_each_column
```

```python
percnt_missing(df)
```

```python
#numbers of columns have missing data
def colunm_missing(df):
    percentage_NAN_in_each_column=percnt_missing(df)
    count_NaN=0
    for index in df:
        if percentage_NAN_in_each_column[index]>0:
            count_NaN+=1        
    print("The number of columns that contain NAN :",count_NaN)
colunm_missing(df)
```

```python
# Delete columns which contain more than 50% of NaN values
def cleanBigNaN(df,percnt_missing):
    number_valid =  int(((100-percnt_missing)/100)*row_number + 1)
    df.dropna( axis=1, thresh=number_valid,inplace=True)
    return df
df = cleanBigNaN(df,50)
df.shape
```

```python
colunm_missing(df)
```

```python
#fill missing values by global constant(((numbers = 999 & string = Missing)))
#after I understand that the data frame has an identifier feature, so I run this code before filling in the interpolation
df= df.fillna({'posteam':'Missing',
                   'posteam_type':'Missing',
                   'defteam':'Missing',
                  'side_of_field':'Missing',
                   'time':999,
                  'yrdln':'Missing',
                   'play_type':'Missing',
                   'pass_length':'Missing',
                   'pass_location':'Missing',
                  'run_location':'Missing',
                   'passer_player_id':999,
                   'passer_player_name':'Missing',
                   'receiver_player_id':999,
                   'receiver_player_name':'Missing',
                   'rusher_player_id':999,
                   'rusher_player_name':'Missing',
                   'solo_tackle_1_team':'Missing',
                   'solo_tackle_1_player_id':999,
                   'solo_tackle_1_player_name':'Missing',
                  })
```

```python
colunm_missing(df)
```

```python
#fill missing values by interpolate
df=df.interpolate(method='linear')
```
